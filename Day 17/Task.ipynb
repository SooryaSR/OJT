```python
#1. Convert 1D array with 8 elements to 3D array with 2x2 elements

import numpy as np

array_1d = np.array([1, 2, 3, 4, 5, 6, 7, 8])  #1D array with 8 elements
array_3d = array_1d.reshape((2, 2, 2))         #Reshape the array
print(array_3d)
```

```python
#2. Convert the array into a 1D array

import numpy as np
array_3d = np.array([[[1, 2], [3, 4]], 
                     [[5, 6], [7, 8]]]) #Create a 3D array 

array_1d = array_3d.reshape(-1)         #Reshape the 3D to 1D array

print(array_1d)
```

```python
#3. Use the scatter() method to draw a scatter plot diagram
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([2, 3, 4, 5, 6, 7, 8, 9])

plt.scatter(x, y)

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(1000)
y = np.random.rand(1000)

plt.scatter(x, y)

plt.title('Scatter Plot with 1000 Dots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 7, 11])

model = LinearRegression()
model.fit(x, y)

x_new = np.array([[6]])
y_pred = model.predict(x_new)

print(f"Predicted value of y when x = 6: {y_pred[0]}")
```
